```python
import pandas as pd
import numpy as np
#
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
#
from tabpfn import TabPFNClassifier
#
from sklearn.preprocessing import LabelEncoder
```

```python
dataset = pd.read_csv('spotify-tracks-dataset/dataset.csv')
dataset.info()
```

```python
dataset.head()
```

```python
# dropeamos la primera columna que no nos aporta nada más que un identificador que tenemos automático 
dataset.drop(['Unnamed: 0', 'track_id'], axis=1, inplace=True)
dataset.info()
```

```python
# para el algoritmo que vamos a utilizar necesitamos todas las columnas con floats
labelEncoder = LabelEncoder()

dataset['artists'] = labelEncoder.fit_transform(dataset['artists'])
dataset['album_name'] = labelEncoder.fit_transform(dataset['album_name'])
dataset['track_name'] = labelEncoder.fit_transform(dataset['track_name'])
dataset['explicit'] = labelEncoder.fit_transform(dataset['explicit'])
dataset['track_genre'] = labelEncoder.fit_transform(dataset['track_genre'])
```

```python
dataset.head()
```

```python
# Eliminamos los valores na (non-defined)
dataset.dropna(inplace=True)
dataset.info()
```

```python
# Separamos en entrenamiento y testeo 
X_train, X_test, y_train, y_test = train_test_split(dataset.drop('track_name', axis=1), 
                                                    dataset['track_name'], test_size=0.33, random_state=42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Creamos el clasificador
clasificador = TabPFNClassifier(device='cpu', N_ensemble_configurations=32)
clasificador.fit(X_train, y_train)
```
